```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
soup = BeautifulSoup(data.text)
standings_table = soup.select('table.stats_table')[0]
```

```python
links = [l.get("href") for l in standings_table.find_all('a')]
links = [l for l in links if '/squads/' in l]
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
data = requests.get(team_urls[0])
matches = pd.read_html(data.text, match="Scores & Fixtures")
```

```python
soup = BeautifulSoup(data.text)
links = [l.get("href") for l in soup.find_all('a')]
print(links)
goalkeeping_link = [l for l in links if l and 'all_comps/keeper/' in l]
print(goalkeeping_link)
```

```python

goalkeeping_data = requests.get(f"https://fbref.com{goalkeeping_link[0]}")
```

```python

goalkeeping = pd.read_html(goalkeeping_data.text, match="Goalkeeping")[0]
```

```python
goalkeeping
```

```python
goalkeeping.columns = goalkeeping.columns.droplevel()
```

```python
goalkeeping
```

```python
columns_to_keep = ['Date', 'Time', 'Comp', 'Round', 'Venue', 'Opponent', 'SoTA']
filtered_goalkeeping = goalkeeping[columns_to_keep]
filtered_goalkeeping = filtered_goalkeeping.dropna()
```

```python
filtered_goalkeeping
```

```python
passing_link = [l for l in links if l and 'all_comps/passing/' in l]
passing_data = requests.get(f"https://fbref.com{passing_link[0]}")
passing = pd.read_html(passing_data.text, match="Passing")[0]
```

```python
passing
```

```python
passing.columns.get_level_values(0)
```

```python
passing_filtered = passing.iloc[:, :15]
```

```python
passing_filtered
```

```python
passing_filtered = passing_filtered.dropna()
filtered_goalkeeping = filtered_goalkeeping.dropna()
passing_filtered.columns = passing_filtered.columns.droplevel()
```

```python
passing_filtered
```

```python
#merge 2 dataframes

merged = filtered_goalkeeping.merge(passing_filtered[['Date', 'Cmp', 'Att', 'TotDist', 'PrgDist']], on='Date')
```

```python
merged
```

```python
matches[0]
```

```python
merged2 = merged.merge(matches[0][['Date', 'Day', 'Result', 'GF', 'GA', 'xG', 'xGA', 'Poss', 'Attendance']], on='Date').dropna()
```

```python
merged2.to_csv('test1.csv', index=False)
```

```python
shooting_link = [l for l in links if l and 'all_comps/shooting/' in l]
print(shooting_link)
```

```python
shooting_data = requests.get(f"https://fbref.com{shooting_link[0]}")
shooting = pd.read_html(shooting_data.text, match="Shooting")[0]
```

```python
shooting.columns = shooting.columns.droplevel()
```

```python
merged3 = merged2.merge(shooting[['Date', 'Sh', 'SoT', 'Dist']], on='Date').dropna()
```

```python
merged3.to_csv('test1.csv', index=False)
```

```python
gsc_link = [l for l in links if l and 'all_comps/gca/' in l]
gsc_data = requests.get(f"https://fbref.com{gsc_link[0]}")
gsc = pd.read_html(gsc_data.text, match="GCA")[0]
```

```python
gsc.columns = gsc.columns.droplevel()
```

```python
merged4 = merged3.merge(gsc[['Date', 'SCA', 'GCA']], on='Date').dropna()
```
